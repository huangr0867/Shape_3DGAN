```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import os
```

```python
def generate_sphere(npoints, radius, ndim=3):
    vec = np.random.randn(ndim, npoints)
    vec /= np.linalg.norm(vec, axis=0)
    scaled_vec = np.multiply(vec, radius)
    return scaled_vec

def convert_coord_to_3d(x_data, y_data, z_data, r):
    shape = (r*2,r*2,r*2)
    three_d_array = np.zeros(shape)
    for i in range(len(x_data)):
        three_d_array[round(x_data[i]+(r-1))][round(y_data[i]+(r-1))][round(z_data[i]+(r-1))] = 1
    return three_d_array.astype(int)
```

```python
radius = 15
xi, yi, zi = generate_sphere(1000,radius)
print(convert_coord_to_3d(xi, yi, zi, radius).shape)

xi,yi,zi = convert_coord_to_3d(xi, yi, zi, radius).nonzero()
fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
ax.scatter(xi, yi, zi, c='b', zorder=10)
```
